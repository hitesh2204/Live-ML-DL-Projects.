### Feature Engineering.

```python
#### creating the Date/Month/Year seperate column from Date_of_Journey columns.
big_df['Date']=big_df['Date_of_Journey'].str.split('/').str[0]
big_df['Month']=big_df['Date_of_Journey'].str.split('/').str[1]
big_df['Year']=big_df['Date_of_Journey'].str.split('/').str[2]
```

```python
big_df.head()
```

```python
###Converting all new created column into int datatype.
big_df['Date']=big_df['Date'].astype(int)
big_df['Month']=big_df['Month'].astype(int)
big_df['Year']=big_df['Year'].astype(int)
```

```python
big_df.dtypes
```

```python
### dropping Date_of_Journey column
big_df=big_df.drop(['Date_of_Journey'],axis=1)
big_df.head()
```

```python
## retrive imp data from Arrival_Time colummn.
big_df['Arrival_Time']=big_df['Arrival_Time'].str.split(' ').str[0]
```

```python
big_df.head()
```

```python
### chekcing null vallues in dataset.
big_df.isnull().sum()
```

```python
### Handling null values in Total_Stops.
big_df[big_df['Total_Stops'].isnull()]
```

```python
big_df['Total_Stops']=big_df['Total_Stops'].fillna('1 stop')
```

```python
big_df['Total_Stops']=big_df['Total_Stops'].replace('non-stop','0 stop')
```

```python
big_df.head()
```

```python
big_df['Stop']=big_df['Total_Stops'].str.split(' ').str[0]
```

```python
big_df.head()
```

```python
big_df=big_df.drop(['Total_Stops'],axis=1)
```

```python
big_df.head()
```

```python
big_df['Stop']=big_df['Stop'].astype(int)
```

```python
big_df.dtypes
```

```python
#### creating the Hours/Minutes seperate column from Arrival_Time columns.
big_df['Hours']=big_df['Arrival_Time'].str.split(':').str[0]
big_df['Minutes']=big_df['Arrival_Time'].str.split(':').str[1]
```

```python
big_df.head()
```

```python
big_df=big_df.drop(['Arrival_Time'],axis=1)
big_df.head()
```

```python
#### creating the Hours/Minutes seperate column from Dep_Time columns.
big_df['Departure_Hours']=big_df['Dep_Time'].str.split(':').str[0]
big_df['Departure_Minutes']=big_df['Dep_Time'].str.split(':').str[1]
```

```python
big_df['Departure_Hours']=big_df['Departure_Hours'].astype(int)
big_df['Departure_Minutes']=big_df['Departure_Minutes'].astype(int)
big_df=big_df.drop(['Dep_Time'],axis=1)
big_df.head()
```

```python
big_df['Hours']=big_df['Hours'].astype(int)
big_df['Minutes']=big_df['Minutes'].astype(int)
```

```python
big_df.dtypes
```

```python
#### creating the different route seperate column from Route columns.
big_df['Route 1']=big_df['Route'].str.split(' →').str[0]
big_df['Route 2']=big_df['Route'].str.split(' →').str[1]
big_df['Route 3']=big_df['Route'].str.split(' →').str[2]
big_df['Route 4']=big_df['Route'].str.split(' →').str[3]
big_df['Route 5']=big_df['Route'].str.split(' →').str[4]
big_df=big_df.drop(['Route'],axis=1)
big_df.head()
```

```python
## Handling null values in columns.
big_df['Price'].fillna(big_df['Price'].mean(),inplace=True)

big_df['Route 1'].fillna('None',inplace=True)
big_df['Route 2'].fillna('None',inplace=True)
big_df['Route 3'].fillna('None',inplace=True)
big_df['Route 4'].fillna('None',inplace=True)
big_df['Route 5'].fillna('None',inplace=True)
```

```python
big_df.isnull().sum()
```

```python
big_df=big_df.drop(['Duration'],axis=1)
big_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
ecoder=LabelEncoder()
big_df['Airline']=ecoder.fit_transform(big_df['Airline'])
big_df['Source']=ecoder.fit_transform(big_df['Source'])
big_df['Destination']=ecoder.fit_transform(big_df['Destination'])
big_df['Additional_Info']=ecoder.fit_transform(big_df['Additional_Info'])

big_df['Route 1']=ecoder.fit_transform(big_df['Route 1'])
big_df['Route 2']=ecoder.fit_transform(big_df['Route 2'])
big_df['Route 3']=ecoder.fit_transform(big_df['Route 3'])
big_df['Route 4']=ecoder.fit_transform(big_df['Route 4'])
big_df['Route 5']=ecoder.fit_transform(big_df['Route 5'])
```

### RandomForestRegressor.

```python
from sklearn.model_selection import RandomizedSearchCV

## number of trees in random forest.
n_estimators=[int(x) for x in np.linspace(start=100,stop=1000,num=12)]
## number of feature in evey split.
max_features=['auto','sqrt']
## maximum number of level in trees.
max_depth=[int(x) for x in np.linspace(5,30,num=6)]
## minimum sample split.
min_samples_split=[2,5,10,15,100]
## min number of sample leaf node.
min_samples_leaf=[1,2,5,10]
```

```python
## creating random grid.
random_grid={'n_estimators':n_estimators,
             'max_features':max_features,
             'max_depth':max_depth,
             'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
## creating RandomForestRegressor.
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=50,cv=5,verbose=2,random_state=0)
```

```python
rf_random.fit(x_train,y_train)
```
