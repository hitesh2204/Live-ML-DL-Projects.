```python
### importing important libraries.
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Sentence.
sent=['The glass of milk',
      'the glass of juice',
      'the cup of tea',
      'i am good boy',
      'I am good developer',
      'understand the meaning of word',
      'your videos are good']
```

```python
sent
```

```python
### Initializing the vocab size.
voc_size=10000
```

### One hot Representation

```python
onehot_rep=[one_hot(words,voc_size)for words in sent]
print(onehot_rep)
```

### Word Embedding Presentation.

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
sent_length=8
embedded_doc=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
### proving the dimention.
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
### printing  values in sparx matrix.
print(model.predict(embedded_doc))
```
