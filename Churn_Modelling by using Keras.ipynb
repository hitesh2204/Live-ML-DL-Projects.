```python
### Importing necessoy libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Reading dataset.
churn_df=pd.read_csv("E://Datasets//New Dataset//Churn_Modelling.csv")
churn_df.head()
```

```python
###chechinf shape of data
churn_df.shape
```

```python
### checking null values in dataset.
churn_df.isnull().sum()
```

```python
### Splitting data inot independent and dependent vaiable.
x_data=churn_df.iloc[:,3:13]
y_data=churn_df.iloc[:,-1]
x_data.head()
```

```python
y_data.head()
```

```python
### coonverting categorical data into numeric form.
Geography=pd.get_dummies(x_data["Geography"],drop_first=True)
Gender=pd.get_dummies(x_data["Gender"],drop_first=True)
```

```python
Gender.head()
```

```python
x_data=pd.concat([x_data,Geography,Gender],axis=1)
```

```python
x_data.head()
```

```python
### Dropping categorical columns.
X_data=x_data.drop(['Geography','Gender'],axis=1,inplace=True)
x_data.head()
```

```python
###splitting data into training an testing dataset.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.25,random_state=0)
```

```python
### convert data into stadard scaler using standard scaler library.
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

### Creating ANN Model

```python
### Impotring impotant libaries.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
### Initializing the ANN.
model=Sequential()

### Addding input layer and first hidden layer.
model.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))

### Addding input layer and second hidden layer.
model.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))

### Addding output layer.
model.add(Dense(units=6,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
### Compiling ANN  model.
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
### Fitting ANN on training data.
model.fit(x_train,y_train,validation_split=0.30,batch_size=10,epochs=100)
```

```python
### predicting test result.
y_pread=model.predict(x_test)
#y_pread=(y_pread>0.5)
print(y_pread)
```
