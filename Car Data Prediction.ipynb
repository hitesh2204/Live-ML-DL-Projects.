```python
## importing libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Finding the unique values in all categorical variable.
print(car_df['Fuel_Type'].unique())
print(car_df['Seller_Type'].unique())
print(car_df['Transmission'].unique())
print(car_df['Owner'].unique())
```

```python
final_dataset=car_df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_dataset.head()
```

```python
final_dataset['current_year']=2020
```

```python
final_dataset.head()
```

```python
final_dataset['no_of_year']=final_dataset['current_year']-final_dataset['Year']
```

```python
final_dataset.head()
```

```python
## dropping unused columns form dataset.
final_dataset.drop(['Year','current_year'],axis=1,inplace=True)
```

```python
final_dataset.head()
```

```python
### Converting categorical data into numeric form.
final_dataset=pd.get_dummies(final_dataset,drop_first=True)
```

```python
final_dataset.head()
```

```python
### FInding the coelation in each feature with other features..
final_dataset.corr()
```

```python
### Visualizing the coelation in graph.
import seaborn as sns
```

```python
sns.pairplot(final_dataset)
```

```python
corrmat=final_dataset.corr()
top_corr_feature=corrmat.index
plt.figure(figsize=(10,10))
sns.heatmap(final_dataset[top_corr_feature].corr(),annot=True,cmap='RdYlGn')
```

```python
### Creating the Independent and Dependent feature.
X=final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
### Finding the most impotant features.
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
feature_imp=pd.Series(model.feature_importances_,index=X.columns)
feature_imp.nlargest(5).plot(kind='barh')
plt.show()
```

```python
### splitting data into taining and testing dataset.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
### Hyperpametrs Tunning.
from sklearn.model_selection import RandomizedSearchCV

## number of trees in random forest.
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
## number of feature in evey split.
max_features=['auto','sqrt']
## maximum number of level in trees.
max_depth=[int(x) for x in np.linspace(5,30,num=6)]
## minimum sample split.
min_samples_split=[2,5,10,15,100]
## min number of sample leaf node.
min_samples_leaf=[1,2,5,10]
```

```python
random_grid={'n_estimators':n_estimators,
             'max_features':max_features,
             'max_depth':max_depth,
             'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
rf=RandomizedSearchCV(estimator=rf_random,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=50,cv=5,verbose=2,random_state=0)
```

```python
rf.fit(X,y)
```

```python
rf_random=RandomForestRegressor(n_estimators=100,max_depth=10,max_features='auto',min_samples_leaf=5,min_samples_split=10)
```

```python
rf_random.fit(X,y)
```

```python
prediction=rf_random.predict(x_test)
```
